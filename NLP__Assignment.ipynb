```python
import nltk
import string
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# 2. INPUT: 2 Paragraphs of Current News (March 10, 2026)
news_text = """
The United Nations nuclear agency is holding an extraordinary meeting this week following recent military strikes in West Asia that have impacted regional stability.
Meanwhile, the global NLP market is projected to reach $67.8 billion by the end of this year as industries rapidly adopt AI-driven text analysis tools.

In other developments, India has secured its third Men's T20 World Cup title after a thrilling victory over New Zealand at the Eden Gardens.
At the same time, experts are warning of a hotter-than-normal summer across the subcontinent, with heatwaves expected to exceed seasonal averages in March.
"""

print("--- Original News Text ---")
print(news_text)
```

```python
# This creates a map to replace all punctuation with an empty string
translator = str.maketrans('', '', string.punctuation)
clean_text = news_text.translate(translator)

print("\n--- After Punctuation Removal ---")
print(clean_text)
```

```python
word_tokens = word_tokenize(clean_text)
sent_tokens = sent_tokenize(news_text) # Done on original to keep sentence structure
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [w for w in word_tokens if w.lower() not in stop_words]

print("\n--- Filtered Tokens (No Stop Words) ---")
print(filtered_tokens)
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

print("\n--- Stemmed Words ---")
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

def get_pos(word):
    """Helper to map NLTK tags to WordNet tags"""
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

lemmatized_words = [lemmatizer.lemmatize(word, get_pos(word)) for word in filtered_tokens]

print("\n--- Lemmatized Words ---")
print(lemmatized_words)
```
