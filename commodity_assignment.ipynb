```python
# Database of commodities and their current rates
commodity_prices = {
    "wheat": "2500 per quintal",
    "rice": "3200 per quintal",
    "cotton": "7000 per quintal",
    "gold": "72000 per 10g"
}
```

```python
def get_commodity_name(user_input):
    user_input = user_input.lower()
    for commodity in commodity_prices.keys():
        if commodity in user_input:
            return commodity
    return None
```

```python
def commodity_name():
    print(" Commodity Price Assistant ")
    print("(Type 'exit' to quit)")

    while True:
        user_input = input("\nUser: ")
        commodity = get_commodity_name(user_input)

        if commodity == "wheat":
            price = commodity_prices["wheat"]
            print(f" The commodity is {commodity} and its price is {price}.")

        elif commodity == "rice":
            price = commodity_prices["rice"]
            print(f" The commodity is {commodity} and its price is {price}.")

        elif commodity == "cotton":
            price = commodity_prices["cotton"]
            print(f" The commodity is {commodity} and its price is {price}.")

        elif commodity == "gold":
            price = commodity_prices["gold"]
            print(f" The commodity is {commodity} and its price is {price}.")

        else:
            # Handle unknown inputs
            print(" I could not find that commodity. Please try wheat, rice, cotton, or gold.")

        if user_input == "exit":
            break
```

```python
commodity_name()
```
